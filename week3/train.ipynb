```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn import datasets
import matplotlib.pyplot as plt
from LogisticRegression import LogisticRegression

X_train = pd.read_csv("D:\\SEM - 6\\MLDL\\ML\\ML-Cardio-Prediction\\data\\X_train_final.csv")
X_test = pd.read_csv("D:\\SEM - 6\\MLDL\\ML\\ML-Cardio-Prediction\\data\\X_test_final.csv")
y_train = pd.read_csv("D:\\SEM - 6\\MLDL\\ML\\ML-Cardio-Prediction\\data\\y_train_final.csv").values.ravel()
y_test = pd.read_csv("D:\\SEM - 6\\MLDL\\ML\\ML-Cardio-Prediction\\data\\y_test_final.csv").values.ravel()

clf = LogisticRegression(lr=0.01)
clf.fit(X_train,y_train)
y_pred = clf.predict(X_test)

def accuracy(y_pred, y_test):
    return np.sum(y_pred==y_test)/len(y_test)

acc = accuracy(y_pred, y_test)
print(acc)
```
