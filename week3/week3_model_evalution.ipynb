# MODEL EVALUATION

Logistic Regression using Sklearn

```python
import numpy as np
import pandas as pd
```

Load Preprocessed Data

```python
X_train = pd.read_csv("D:\\SEM - 6\\MLDL\\ML\\ML-Cardio-Prediction\\data\\X_train_final.csv")
X_test = pd.read_csv("D:\\SEM - 6\\MLDL\\ML\\ML-Cardio-Prediction\\data\\X_test_final.csv")
y_train = pd.read_csv("D:\\SEM - 6\\MLDL\\ML\\ML-Cardio-Prediction\\data\\y_train_final.csv").values.ravel()
y_test = pd.read_csv("D:\\SEM - 6\\MLDL\\ML\\ML-Cardio-Prediction\\data\\y_test_final.csv").values.ravel()

print( X_train.shape, X_test.shape)
```

Train Sklearn Model

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

model = LogisticRegression(max_iter=1000)
model.fit(X_train, y_train)
```

Evaluate Sklearn Model

```python
y_pred = model.predict(X_test)

print("Accuracy:", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

Confusion Matrix

```python
import seaborn as sns
import matplotlib.pyplot as plt

cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot=True, fmt='d')
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()
```

```python
def accuracy(y_true, y_pred):
    return np.mean(y_true == y_pred)

print("Custom Accuracy:", accuracy(y_test, y_pred)*100)
```
