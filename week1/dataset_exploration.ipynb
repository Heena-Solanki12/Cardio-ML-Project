# Identify Problems / Observations


```python
# Outliers
# Check height/weight/blood pressure values
print("Outlier Check:")
plt.figure(figsize=(15, 5))
plt.subplot(1, 3, 1)
sns.boxplot(x=df['height'])
plt.title('Height Distribution with Outliers')
plt.subplot(1, 3, 2)
sns.boxplot(x=df['weight'])
plt.title('Weight Distribution with Outliers')
plt.subplot(1, 3, 3)
sns.boxplot(x=df['ap_hi'])
plt.title('Systolic Blood Pressure Distribution with Outliers')
plt.tight_layout()
plt.show()
```

```python
# Target Distribution (CVD)
plt.figure(figsize=(6, 4))
sns.countplot(x='cardio', data=df)
plt.title('Cardiovascular Disease Distribution')
plt.show()
```

```python
# Correlation Matrix
plt.figure(figsize=(8, 6))
corr = df.corr()
sns.heatmap(corr, annot=True, fmt=".2f", cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```
